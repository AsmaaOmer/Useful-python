# Basic line plot

```python
theta = np.linspace(-9 * np.pi, 9 * np.pi, 300)
s = 3*np.cos(theta)
t = 3*np.sin(theta)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot(s, t, theta)
 
ax.view_init(elev=20, azim=60)
ax.dist = 11
plt.show()
```
