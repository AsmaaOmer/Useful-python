```python
data_path = "/Users/robincole/Documents/Github/mindsdb/docs/examples/time_series/fuel.csv"
df = pd.read_csv(data_path)
df.head()
```

```python
from mindsdb import *


MindsDB().learn(
    predict = 'Main_Engine_Fuel_Consumption_MT_day',
    from_data = df,
    model_name='fuel',

    # Time series arguments:

    order_by='Time',
    group_by='id',
    window_size=24, # just 24 hours

)
```
