```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('darksky_temperature.csv', index_col=0) 
df.index = pd.to_datetime(df.index)  # Conver index to timestamp
df.head()
```

Lets look at a noisy few days

```python
start = '2018-2-15'
end = '2018-2-16'
plt.figure(figsize=(20,10))

plt.plot(df.loc[start:end], 'r', label='raw');
plt.plot(pd.rolling_mean(df.loc[start:end], 100), 'b--', label='rolling_mean');
plt.legend();
```
