Usage from https://github.com/jeffbass/imagezmq

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from imutils.video import WebcamVideoStream
from imutils import opencv2matplotlib
import cv2
```

```python
camera = WebcamVideoStream().start()
```

```python
frame = camera.read()
```

```python
type(frame)
```

```python
plt.imshow(frame)
```

imutils has convenience for fixing BGR -> GBB

```python
image = opencv2matplotlib(frame)
```

```python
plt.imshow(image)
```

```python
type(image)
```

```python
cv2.imshow('camera', image) # display in window
cv2.waitKey(1)
```

```python
camera.stop() # doesn't do anything, just kill kernel to release
```
