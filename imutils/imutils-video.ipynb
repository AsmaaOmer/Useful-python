Usage from https://github.com/jeffbass/imagezmq

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from imutils.video import VideoStream
from imutils import opencv2matplotlib
```

```python
camera = VideoStream().start()
```

```python
image = camera.read()
```

```python
type(image)
```

```python
plt.imshow(image)
```

imutils has convenience for fixing BGR -> GBB

```python
plt.imshow(opencv2matplotlib(image))
```

```python
camera.stop() # doesn't do anything, just kill kernel to release
```
