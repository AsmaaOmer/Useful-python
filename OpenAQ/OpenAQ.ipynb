```python
import openaq

api = openaq.OpenAQ()

status, resp = api.cities(limit=10000)
```

```python
countries = []
for result in resp['results']:
    countries.append(result['country'])
```

```python
set(countries)
```

```python
result, britain = api.locations(country="GB")
```

```python
cities = []
for location in britain['results']:
    cities.append(location['city'])

set(cities)
```

```python
london = []
for location in britain['results']:
    if location['city'] == 'London':
        london.append(location)
        
london
```

```python
api.latest(city="London", location='London Westminster', df=True)
```

```python
wesminster_pm25 = api.measurements(city="London", location='London Westminster', parameter='pm25', limit=10000, df=True)
```
