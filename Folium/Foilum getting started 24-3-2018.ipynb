Folium makes it easy to visualize data thatâ€™s been manipulated in Python on an interactive Leaflet map. It enables both the binding of data to a map for choropleth visualizations as well as passing Vincent/Vega visualizations as markers on the map.

```python
import folium
from folium.plugins import HeatMap
import os
import numpy as np

lat = 51.383578
long = -0.262539
zoom = 14
```

```python
map_osm = folium.Map(location=[lat, long], zoom_start=zoom)
map_osm
```

```python
map_osm.save('map.html')
```

```python
data = (np.random.normal(scale=0.05, size=(100, 3)) *
        np.array([[1, 1, 1]]) +
        np.array([[lat, long, 1]])).tolist()
```

```python
data[:10] # Data is imaginary lat long and reading
```

```python
m = folium.Map([lat, long],  zoom_start=zoom) # tiles='stamentoner',

HeatMap(data).add_to(m)

m
```

```python
m.save('Heatmap.html')
```
