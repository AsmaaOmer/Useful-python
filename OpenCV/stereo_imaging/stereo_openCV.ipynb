The disparity of a pixel is equal to the shift value that leads to minimum sum-of-squared-differences for that pixel.

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline

imgL = cv2.imread('tsukuba_l.png',0)
imgR = cv2.imread('tsukuba_r.png',0)
```

```python
cv2.__version__
```

```python
plt.imshow(imgL)
```

```python
plt.imshow(imgR)
```

Visualise the difference between the images

```python
plt.imshow(imgR - imgL)
```

Now compute the disparity and visualise

```python
stereo = cv2.StereoBM_create(numDisparities=16, blockSize=15)
disparity = stereo.compute(imgL,imgR)
plt.imshow(disparity,'gray')
plt.show()
```
