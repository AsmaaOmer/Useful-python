```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
```

```python
shape = (100, 100)
img_array = np.zeros(shape)

mean = 0
gauss = np.random.normal(mean, 1, img_array.shape)

# normalize image to range [0,255]
noisy = img_array + gauss
minv = np.amin(noisy)
maxv = np.amax(noisy)
noisy = (255 * (noisy - minv) / (maxv - minv)).astype(np.uint8)
```

```python
plt.imshow(noisy);
```

Convert to a PIL image, then save

```python
img = Image.fromarray(noisy)
```

```python
type(img)
```

```python
# img.show() # open in standalone window
```

```python
img.save('img.jpg')
```
